## Отбор компонент

```python
import numpy as np
import pandas as pd
import sklearn.linear_model as skl
```

```python
df = pd.read_csv('insurance.csv')
df = pd.get_dummies(df, prefix=['sex', 'smoker', 'region'])
df.head()
```

### Решение регрессионной задачи

```python
X = df.drop(columns = ['charges']).to_numpy()
y = df[['charges']].to_numpy()

model = skl.LinearRegression()
model.fit(X,y)
model.score(X,y)
```

### Комбинаторное исключение компонент

```python
### Исключение одного признака
cols = list(df.drop(columns = ['charges']).columns)

for i in range(len(cols)):
    X = df.drop(columns = ['charges', cols[i]]).to_numpy()
    y = df[['charges']].to_numpy()
    model = skl.LinearRegression()
    model.fit(X,y)
    print("исключение признака '", cols[i], "': ", model.score(X,y), ) 
```

```python
### Исключение двух признаков
for i in range(len(cols)):
    for j in range(i+1, len(cols)):
        X = df.drop(columns = ['charges', cols[i], cols[j]]).to_numpy()
        y = df[['charges']].to_numpy()
        model = skl.LinearRegression()
        model.fit(X,y)
        print("исключение признаки '", cols[i], "' и '", cols[j], "': ", model.score(X,y), ) 
```

```python
### Исключение трех признаков
for i in range(len(cols)):
    for j in range(i+1, len(cols)):
        for k in range(j+1, len(cols)):
            X = df.drop(columns = ['charges', cols[i], cols[j], cols[k]]).to_numpy()
            y = df[['charges']].to_numpy()
            model = skl.LinearRegression()
            model.fit(X,y)
            print("исключение признаки '", cols[i], "', '", cols[j], "' и '", cols[k], "': ", model.score(X,y), ) 
```

```python
### Исключение четырех признаков
for i in range(len(cols)):
    for j in range(i+1, len(cols)):
        for k in range(j+1, len(cols)):
            for l in range(k+1, len(cols)):
                X = df.drop(columns = ['charges', cols[i], cols[j], cols[k], cols[l]]).to_numpy()
                y = df[['charges']].to_numpy()
                model = skl.LinearRegression()
                model.fit(X,y)
                print("исключение признаки '", cols[i], "', '", cols[j], "', '", cols[k], "' и '", cols[l], "': ", model.score(X,y), ) 
```

```python
### Исключение пяти признаков
for i in range(len(cols)):
    for j in range(i+1, len(cols)):
        for k in range(j+1, len(cols)):
            for l in range(k+1, len(cols)):
                for m in range(l+1, len(cols)):
                    X = df.drop(columns = ['charges', cols[i], cols[j], cols[k], cols[l], cols[m]]).to_numpy()
                    y = df[['charges']].to_numpy()
                    model = skl.LinearRegression()
                    model.fit(X,y)
                    print("исключение признаки '", cols[i], "', '", cols[j], "', '", cols[k], "', '", cols[l], "' и '", cols[m], "': ", model.score(X,y), ) 
```
