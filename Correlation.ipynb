# Корреляция

## Импорт библиотек

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Чтение данных

```python
df = pd.read_csv('ENB2012_data.csv').drop(columns=['Y1','Y2','X2','X3','X4'])
#df = pd.get_dummies(df, prefix=['sex', 'smoker', 'region']).drop(columns='charges')
df.head()
```

### Корреляционный анализ данных

```python
df.corr()
```

### Тепловые карты корреляций

```python
sns.heatmap(df.corr())
```

## Связь между зависимостью и корреляцией

```python
df.describe()
```

```python
plt.figure(figsize=(7,7))

plt.xlim([-5,8])
plt.ylim([-5,8])
plt.grid()

plt.plot(0,0, 'o', color = 'black')
plt.plot(np.arange(-5,9), 0*np.arange(-5,9), '-', color = 'black' )
plt.plot(np.arange(-5,9)*0, np.arange(-5,9), '-', color = 'black' )

plt.plot(df['X6'], df['X5'], '.')

plt.show()
```

### Комбинаторный перебор

```python
cols = df.columns
```

```python
for i in range(len(cols)):
    for j in range(i, len(cols)):
        
        print('признаки', cols[i], cols[j])
        print('корреляция', df.corr()[cols[i]][cols[j]])

        plt.figure(figsize=(7,7))

        plt.xlim([-5,8])
        plt.ylim([-5,8])
        plt.grid()

        plt.plot(0,0, 'o', color = 'black')
        plt.plot(np.arange(-5,9), 0*np.arange(-5,9), '-', color = 'black' )
        plt.plot(np.arange(-5,9)*0, np.arange(-5,9), '-', color = 'black' )

        plt.plot(df[cols[i]], df[cols[j]], '.')

        plt.show()
```

### Функциональная и корреляционная зависимость

```python
df['X1^2'] = df['X1']**2
df
```

```python
plt.figure(figsize=(7,7))

plt.xlim([-2,2])
plt.ylim([-2,2])
plt.grid()

plt.plot(0,0, 'o', color = 'black')
plt.plot(np.arange(-2,3), 0*np.arange(-2,3), '-', color = 'black' )
plt.plot(np.arange(-2,3)*0, np.arange(-2,3), '-', color = 'black' )

plt.plot(df['X1'], df['X1^2'], '.')

plt.show()
```
